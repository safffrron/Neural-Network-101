```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
dataset = pd.read_csv('../../Datasets/universitydatasets.csv', encoding='unicode_escape')
```

```python
dataset
```

```python
le = LabelEncoder()
dataset["Location"] = le.fit_transform(dataset["Location"])
print(dataset["Location"])
dataset["Institution"] = le.fit_transform(dataset["Institution"])
dataset["Institution"]
```

```python
cols = dataset.columns
dataset[cols] = dataset[cols].apply(pd.to_numeric, errors='coerce')
dataset.update(dataset[cols].fillna(1000))
dataset
```

```python
X = dataset.iloc[ : , :-1]
Y = dataset.iloc[ : , -1]

X = X / X.max(axis=0)
print(X)
print(Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
print(model.coef_)
model.intercept_
```

```python
Y_pred = model.predict(X_test).flatten()    # predicted value of y_test
#Y_test = Y_test.to_numpy().flatten()

print(Y_pred.shape)
print(Y_test.shape)
print(Y_test)
print(Y_pred)
```

```python
metrics.r2_score(Y_test,Y_pred)
```

```python
score = model.score(X_test, Y_test)
print(score)
```
